```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.svm import SVC
from sklearn.datasets import load_iris
from sklearn.model_selection  import train_test_split
```

```python
iris = load_iris()
```

```python
X = iris.data
Y = iris.target
```

# Principal Component Analysis

```python
from sklearn.decomposition import PCA
```

```python
X_pca = PCA(n_components=2)
reduced_data_X_pca = X_pca.fit_transform(X)
print('reduced_data_X_pca.shape :', reduced_data_X_pca.shape)
print('reduced_data_X_pca :')
print(reduced_data_X_pca)
```

# Machine learning

```python
from sklearn.model_selection import train_test_split
```

```python
X = reduced_data_X_pca
```

```python
plt.scatter(X[:,0], X[:,1], c = Y, cmap = 'Paired')
```

```python
clf = SVC(gamma = 'auto')
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
clf.fit(x_train, y_train)
y_predict = clf.predict(x_test)
y_predict - y_test 
```

```python
plt.scatter(x_test[:,0], x_test[:,1], c = y_predict - y_test)
```

```python
x0 = np.arange(X.min() - 0.1, X.max(), 0.02)  # 圖比較好看
y0 = np.arange(X.min(), X.max(), 0.02)
xm, ym = np.meshgrid(x0, y0)
P = np.c_[xm.ravel(), ym.ravel()]
z = clf.predict(P)
Z = z.reshape(xm.shape)
plt.contourf(xm, ym, Z, cmap = 'Paired', alpha = 0.2)
plt.scatter(x_test[:,0], x_test[:,1], cmap = 'Paired', c = y_test)
```
