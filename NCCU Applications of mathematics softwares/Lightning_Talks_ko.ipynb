```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.linspace(-10, 10, 1000)
y = x*x*x*x - 3*x*x*x + 5000*np.sin(x)
plt.figure(figsize = (20,10))
plt.plot(x, y)
plt.title('$x^4 -3x^3 + 5000sin(x)$的圖形', c = 'r', fontsize = 20)
plt.xlabel('-10到10範圍', c = 'r', fontsize = 20)
plt.ylabel('$x^4 -3x^3 + 5000sin(x)$', c = 'r', fontsize = 20)
plt.legend(y, labels = ['可以顯示中文了！！'] , loc = 'upper center', fontsize = 20)
plt.tick_params(axis ='both')
```
