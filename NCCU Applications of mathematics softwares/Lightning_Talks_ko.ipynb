```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.linspace(-10, 10, 1000)
y = x*x*x*x - 3*x*x*x + 5000*np.sin(x)
plt.figure(figsize = (20,10))
plt.plot(x, y)
plt.title('$x^4 -3x^3 + 5000sin(x)$', c = 'r', fontsize = 20)
plt.xlabel('-1010', c = 'r', fontsize = 20)
plt.ylabel('-500015000', c = 'r', fontsize = 20)
plt.legend(y, labels = [''] , loc = 'upper center', fontsize = 20)
plt.tick_params(axis ='both')
```
